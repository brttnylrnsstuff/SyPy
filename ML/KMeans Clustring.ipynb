### KMeans Clustring 

unsupervised learning algo. it takes unlabelled data, it attempts to group similar data points in a cluster
in this data points are assigned to new cluster until formed centroid and no class remains

elbow method: determine value of k based on sse, sum of square error is a distance between data points 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data
```

```python
data[0]
data[0].shape
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(data[0])
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize = (10,6))


ax1.set_title('K Means Model')
ax1.scatter(data[0][:, 0], data[0][:,1], c=model.labels_, cmap='rainbow')


ax2.set_title('K Means Orignal Data')
ax2.scatter(data[0][:, 0], data[0][:,1], c=data[1], cmap='rainbow')
```
