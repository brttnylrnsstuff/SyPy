## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,  5, 11)
y = x **2
```

#### Functional method

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title("Title")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

#### Object Oriented API Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])


axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title("Larger Plot")
axes2.set_title("Smaller Plot")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x,y)
```

```python
axes[0].plot(x,y)
axes[0].set_title("First")

axes[1].plot(y,x)
axes[1].set_title("Second")
```

### Figure size and DPT

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
```

```python
axes[0].plot(x,y)
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
axes.plot(x, x**2, label="X Square")
axes.plot(x, x**3, label="X Cube")

axes.legend()
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
axes.plot(x, x**2, label="X Square")
axes.plot(x, x**3, label="X Cube")

axes.legend(loc=10)
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
axes.plot(x, x**2, label="X Square")
axes.plot(x, x**3, label="X Cube")

axes.legend(loc=10)
axes.legend(loc=(0.1, .1))
```

### Setting colours (Plot Appearence)

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color='orange', linewidth=2)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color='orange', lw=5, marker='*', markersize=20)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color='orange', lw=5, marker='*', markersize=20, markerfacecolor="yellow", markeredgecolor="green")
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color='orange', lw=2, ls='--')
axes.set_xlim([0, 1])
axes.set_ylim([0, 1])
```
