## Seaborn

#### Distribution plots for seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.jointplot(x=tips["total_bill"], y=tips["tip"])
```

```python
sns.pairplot(tips, hue="sex", palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False) #KDE = Kernal density estimation
```

```python
sns.distplot(tips['total_bill'], kde=True, bins=40) 
```

```python
sns.kdeplot(tips['total_bill']) 
```

### Matrix Plots

```python
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white',linewidths=1)
```

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, col='day', row='time', hue='sex', aspect=.6, size=8)
```

### Grids

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

### Style and color

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(right=True, trim=True)
```

```python
plt.figure(figsize=(12,3))
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(right=True, trim=True)
```

```python
plt.figure(figsize=(12,3))
sns.set_style('ticks')
sns.set_context('poster')
sns.countplot(x='sex', data=tips)
sns.despine(right=True, trim=True)
```

```python
plt.figure(figsize=(12,3))
sns.set_style('ticks')
sns.set_context('notebook', font_scale=3)
sns.countplot(x='sex', data=tips)
sns.despine(right=True, trim=True)
```
